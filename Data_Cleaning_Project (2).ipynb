# DATA CLEANING & VISUALIZATION PROJECT
This project focuses on cleaning a raw dataset and extracing meaningful insights using Python libraries like Pandas, Matplotlib and Seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("netflix1.csv.zip")
df.head()
```

## Dataset Description

The dataset used in this project contains information about Netflix shows and movies. It includes details such as title, type, country, release year, rating, and duration.
The dataset consists of multiple records and provides insights into content distribution across different countries and categories.

## Data Cleaning

- Missing values were identified using `.isnull()` function.
- Missing values were handled using forward fill method.
- Duplicate records were removed to ensure data quality.
- Data types were checked and verified.

These steps improved the reliability and consistency of the dataset.

```python
df.info()
df.describe()
df.isnull().sum()
```

## Clean Data

```python
df.drop_duplicates(inplace=True)
df.fillna(method='ffill',inplace=False)
```

## Visualization

## Analysis 1: Content Ratings

- The most common ratings include TV-MA and PG-13.
- This shows that most content is targeted towards mature audiences.
- There is a smaller portion of content for children.

```python
sns.countplot(x='rating',data=df)
plt.show()
```

## Analysis 2: Movies vs TV Shows

- The dataset contains both Movies and TV Shows.
- Movies are more frequent compared to TV Shows.
- This indicates that Netflix has a larger collection of movies.

```python
sns.countplot(x='type',data=df)
plt.show()
```

## Analysis 4: Release Year Distribution

- Most content has been released in recent years.
- There is a noticeable increase in content production after 2015.
- Older content exists but is less frequent.

```python
sns.countplot(x='release_year',data=df)
plt.show()
```

```python
df['release_year'].hist()
```

## Analysis 5: Top 10 Countries by Content

- The United States has the highest number of movies and TV shows in the dataset.
- India is the second-largest contributor to content.
- The United Kingdom and Pakistan also have a notable number of entries.
- Some entries had missing country values labeled as "Not Given".
- Overall, content production is dominated by a few countries, while others contribute less.

```python
top_countries=df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values,y=top_countries.index)
plt.title("Top 10 countries by content")
plt.xlabel("count")
plt.ylabel("Country")
plt.show()
```
